```python
#Q.1 importing labraries & data
import h5py
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

filepath = "faceimages.mat"
arrays={}
f=h5py.File(filepath)
mat={k:v for k, v in f.items() if k[0] != '_'}
mat1=mat['data']
data=pd.DataFrame({k: pd.Series(v[0]) for k, v in mat.items()})
```

```python
data1=pd.DataFrame(mat1).T
data1
#Q.2 Contructed Data Matrix
```

```python
#Q.2 Mean - Centered data
from sklearn.preprocessing import StandardScaler
X = data1.iloc[:, 0:10304].values
y = data1.iloc[:,10304].values
X = StandardScaler().fit_transform(X)

```

```python
#Q.3 Covariance matrix of data
covariance_matrix = np.cov(X.T)
print("covariance matrix", covariance_matrix)
```

```python
#Eigen Values & Vectors from covariance matrix
eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
print("Eigen Values", eigen_values, "\n")
print("Eigen Vectors", eigen_vectors)
```

```python
eigen_vec_svd=np.linalg.svd(X.T)
eigen_vec_svd
```

```python
for val in eigen_values:
    print(val)
```

```python
variance_explained = [(i / sum(eigen_values)) * 100 for i in eigen_values]
variance_explained
```

```python
#Cummilative variance denotes how much the first 'k' component contribute towrds total energy
cumulative_variance_explained = np.cumsum(variance_explained)
cumulative_variance_explained
print("\n", cumulative_variance_explained)
```

```python
#Energy Curve of PCA (First 1000 Principal values, as Cummilative Variance gets saturated to 100% below first 500 Principal Values)
plt.plot(cumulative_variance_explained[0:1000])
plt.xlabel("Number of components")
plt.ylabel("Cumulative explained variance")
plt.title("PCA Energy Curve")
plt.show()
```

```python
#Q.4 Projection Matrix with first 200 Eigen vectors as they combined contribute 95% of energy 
projection_matrix=eigen_vectors[0:200, :]
projection_matrix.T
```

```python
X_pca = X.dot(projection_matrix.T)
X_pca
```

```python
# Projection matrix with 5 principal components
import matplotlib.pyplot as plt
from sklearn import decomposition
pca = decomposition.PCA()
pca.n_components = 5
lowdimdata = pca.fit_transform(X)
lowdimdata_df = pd.DataFrame(data = lowdimdata)
lowdimdata_df
```

```python
# Reconstruction error for 5 principal components
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from math import sqrt
r2 = r2_score(X[:,:5],lowdimdata)
rmse = sqrt(mean_squared_error(X[:,:5],lowdimdata))
rmse
```
